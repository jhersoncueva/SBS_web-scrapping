```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os

# Configuración de la ruta de descarga
download_dir = "../data"  # Ruta relativa desde el script
absolute_download_dir = os.path.abspath(download_dir)  # Convertir a ruta absoluta
if not os.path.exists(absolute_download_dir):
    os.makedirs(absolute_download_dir)  # Crear la carpeta si no existe

# Configuración del navegador y opciones de descarga
options = webdriver.ChromeOptions()
options.add_experimental_option("prefs", {
    "download.default_directory": absolute_download_dir,  # Directorio de descarga
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
})

# Iniciar el navegador con las opciones configuradas
driver = webdriver.Chrome(options=options)

# Navegar a la página principal
url = "https://www.sbs.gob.pe/app/stats_net/stats/EstadisticaBoletinEstadistico.aspx?p=1#"
driver.get(url)
time.sleep(3)  # Esperar a que la página cargue

# Primer click para acceder al menú general (Indicadores Financieros)
driver.find_element(By.XPATH, '//*[@id="Bot_00"]/div/a').click()
time.sleep(2)

# Iterar sobre los elementos de la lista para encontrar "Estados Financieros por Empresa Bancaria"
menu_items = driver.find_elements(By.XPATH, '//*[@id="Bot_00"]/following-sibling::div//a')

for item in menu_items:
    if item.text == "Estados Financieros por Empresa Bancaria":
        item.click()  # Hacer clic cuando se encuentre el enlace correcto
        time.sleep(3)
        break  # Salir del bucle una vez que se ha encontrado y hecho clic en el enlace correcto

# Descargar archivos de la sección "Indicadores Financieros"
def download_excel_files():
    # Seleccionar todos los enlaces de meses visibles en la sección
    months_links = driver.find_elements(By.XPATH, "//div[@class='item-boletin-iframe']//a")
    
    # Hacer clic en cada enlace para descargar el archivo
    for link in months_links:
        file_url = link.get_attribute("href")
        month_name = link.text
        if file_url.endswith(".XLS"):  # Verificar que sea un archivo Excel
            driver.get(file_url)  # Descargar el archivo
            print(f"Descargando archivo de {month_name}: {file_url}")
            time.sleep(2)  # Pausa para asegurar la descarga

# Ejecutar la función de descarga para todos los meses visibles
download_excel_files()

# Cerrar el navegador después de completar las descargas
driver.quit()

print("Descarga completada.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os

# Configuración de la ruta de descarga
download_dir = "../data"  # Ruta relativa desde el script
absolute_download_dir = os.path.abspath(download_dir)  # Convertir a ruta absoluta
print("Ruta de descarga configurada:", absolute_download_dir)  # Verificar la ruta de descarga

if not os.path.exists(absolute_download_dir):
    os.makedirs(absolute_download_dir)  # Crear la carpeta si no existe

# Configuración del navegador y opciones de descarga
options = webdriver.ChromeOptions()
options.add_experimental_option("prefs", {
    "download.default_directory": absolute_download_dir,  # Directorio de descarga
    "download.prompt_for_download": False,
    "download.directory_upgrade": True,
    "safebrowsing.enabled": True
})

# Iniciar el navegador con las opciones configuradas
driver = webdriver.Chrome(options=options)

# Navegar a la página principal
url = "https://www.sbs.gob.pe/app/stats_net/stats/EstadisticaBoletinEstadistico.aspx?p=1#"
driver.get(url)
time.sleep(5)  # Esperar a que la página cargue completamente

# Primer click para acceder al menú general (Boletín Informativo Mensual)
driver.find_element(By.XPATH, '//*[@id="Bot_00"]/div/a').click()
time.sleep(3)

# Navegar y expandir "Estados Financieros por Empresa Bancaria"
menu_items = driver.find_elements(By.XPATH, '//*[@id="Bot_00"]/following-sibling::div//a')
for item in menu_items:
    if item.text == "Estados Financieros por Empresa Bancaria":
        item.click()  # Hacer clic en "Estados Financieros por Empresa Bancaria"
        time.sleep(3)
        break

# Asegurarse de que "Indicadores de las Empresas Bancarias" esté visible y hacer clic
for item in menu_items:
    if item.text == "Indicadores de las Empresas Bancarias":
        item.click()  # Hacer clic en "Indicadores de las Empresas Bancarias"
        time.sleep(3)
        break

# Finalmente, hacer clic en "Indicadores Financieros"
driver.find_element(By.XPATH, '//*[@id="Bot_07"]/div/a').click()
time.sleep(5)

# Descargar archivos de la sección "Indicadores Financieros"
def download_excel_files():
    months_links = driver.find_elements(By.XPATH, "//div[@class='item-boletin-iframe']//a")
    
    # Hacer clic en cada enlace para descargar el archivo
    for link in months_links:
        month_name = link.text
        file_url = link.get_attribute("href")
        print(f"Descargando archivo de {month_name}: {file_url}")
        link.click()  # Clic en el enlace para iniciar la descarga
        time.sleep(15)  # Pausa extendida para permitir la descarga completa de cada archivo

# Ejecutar la función de descarga para todos los meses visibles
download_excel_files()

# Cerrar el navegador después de completar las descargas
driver.quit()

print("Descarga completada.")
```
